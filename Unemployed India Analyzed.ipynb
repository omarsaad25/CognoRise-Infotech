```python
df['Estimated Unemployment Rate (%)']=pd.to_numeric(df['Estimated Unemployment Rate (%)'],errors='coerce')
```

```python
df['Region'] = df['Region'].astype('category')
```

```python
numeric_df = df.select_dtypes(include=[np.number])
```

```python
df = pd.concat([df[['Region', 'Date']], numeric_df], axis=1)
```

```python
grouped = df.groupby(['Region', 'Date']).mean().reset_index()
```

```python
top_regions = grouped['Region'].value_counts().head(10).index
filtered_data = grouped[grouped['Region'].isin(top_regions)]
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_data, x='Date', y='Estimated Unemployment Rate (%)')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'], 'bo-', label='Estimated Unemployment Rate (%)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
```

```python
###from the previos line charts we concluded that the max unemployment was from 2020-03 to 2020-06 due to corona virus
```

```python
heatmap_data = df.pivot_table(values='Estimated Unemployment Rate (%)', index='Region', columns='Date')
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(
    heatmap_data,
    cmap='YlGnBu',
    linewidths=0.5,
    annot=True,
    fmt='.1f', 
    cbar_kws={'label': 'Unemployment Rate (%)'}
)
plt.xticks(rotation=45, ha='right')
plt.title('Unemployment Rate Heatmap', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.tight_layout()

plt.show()
```

```python
### we learned that Sudden spikes during events like COVID-19 are clearly visible showing the widespread impact and showing the high unemployment rate in tripura and hyryana
```

```python
from prophet import Prophet
```

```python
df_prophet = df.rename(columns={'Date': 'ds', 'Estimated Unemployment Rate (%)': 'y'})

model = Prophet()
model.fit(df_prophet)

future = model.make_future_dataframe(periods=12, freq='M')

forecast = model.predict(future)

fig = model.plot(forecast)
plt.title('Forecast of Unemployment Rate')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```
